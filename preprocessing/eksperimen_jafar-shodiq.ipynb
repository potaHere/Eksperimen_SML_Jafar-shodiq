```python
# standarisasi
numerical_column = df.select_dtypes(include='number').columns.tolist()

scaler = StandardScaler()
df_scaled = df.copy()

df_scaled[numerical_column] = scaler.fit_transform(df[numerical_column])

df_scaled
```

```python
# one-hot encoding for only color_category
categorical_column = ['color_category']  # Put only this column in a list

encoder = OneHotEncoder(sparse_output=False)

# Apply one-hot encoding only to the color_category column
encoded_features = encoder.fit_transform(df[categorical_column])
encoded_df = pd.DataFrame(
    encoded_features, 
    columns=encoder.get_feature_names_out(categorical_column),
    index=df.index
)

# Combine with the scaled dataframe
df_encoded = pd.concat([df_scaled, encoded_df], axis=1)

# Optionally, drop the original color_category column
df_encoded = df_encoded.drop('color_category', axis=1)

df_encoded
```

```python
df_encoded.to_csv('avocado_ripeness_dataset_preprocessed.csv', index=False)
```
